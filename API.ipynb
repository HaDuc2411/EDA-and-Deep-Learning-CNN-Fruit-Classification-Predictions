```python
def load_pred_img(filename, img_shape=100):
  imgs = tf.io.read_file(filename)
  imgs = tf.image.decode_image(imgs)
  imgs = tf.image.resize(imgs, size =[100,100])
  imgs = imgs/255.
  return imgs

image = load_pred_img('/content/drive/MyDrive/tf_milestone/Banana_0.jpg')
imgs_backend = tf.expand_dims(image, 0)
imgs_backend = imgs_backend.numpy().tolist()
```

```python
import json

input_data_json = json.dumps({
    'signature_name':'serving_default',
    'instances':imgs_backend})
```

```python
input_data_json
```

```python
import requests

URL = "https://fruit-backend-ilmi.herokuapp.com/v1/models/fruit_classification:predict"
r = requests.post(URL, data=input_data_json)
result =r.json()
pred=result['predictions']
pred
```
